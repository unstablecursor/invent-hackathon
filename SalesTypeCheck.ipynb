```python
import pandas as pd
inventory_positions = pd.DataFrame.from_csv("files/InventoryPosition.csv")
products = pd.DataFrame.from_csv("files/Products.csv")
stores = pd.DataFrame.from_csv("files/Stores.csv")

```

```python
import numpy as np
for product, item in products.iterrows():
    x = inventory_positions.loc[lambda df: df.ProductCode == product, :]
    #print x
    arr = []
    arr2 = []
    for store, item in stores.iterrows():
        #save_location = 'results/' + store + '_' + product + '.csv'
        ress = x.loc[lambda df: df.StoreCode == store, :]
        ress = ress.tail(1)
        arr.append(ress['StoreStock'].values)
        arr2.append(ress['StoreCode'].values)
    arrr = [val for sublist in arr for val in sublist]
    arrr2 = [val for sublist in arr2 for val in sublist]
    print arrr
    print arrr2
    
```

```python
pg = products.loc[lambda df: df.ProductGroup =='PG5', :]
pg
```

```python
from datetime import datetime
now = datetime.strptime("01-31-2017", "%m-%d-%Y")
start = datetime.strptime("01-06-2015", "%m-%d-%Y")

product_series = pd.Series(index=pd.date_range(start, now))
product_series.fillna(0)
```

```python

for product, val in pg.iterrows():
    print inventory_positions.loc[lambda df: df.ProductCode == product, :].groupby('Date')['SalesQuantity'].sum()
    product_series = product_series.add(inventory_positions.loc[lambda df: df.ProductCode == product, :].groupby('Date')['SalesQuantity'].sum(),fill_value=0)
```

```python
product_series = product_series.fillna(0)
```

```python
product_series.to_csv('pg5.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(product_series)
figure = plt.gcf() # get current figure
figure.set_size_inches(32, 24)
# when saving, specify the DPI
plt.savefig("myplot.png", dpi = 100)
#plt.savefig('pg1.png', format='png', dpi=1200)
plt.show()
```
